```python
import pandas as pd
```

```python
df=pd.read_csv("netflix_data.csv")
```

```python
df.duplicated()
```

```python
 df.isnull().sum()
```

```python
 df=df.dropna()
```

```python
df
```

```python
 df=df.sort_values("release_year",ascending=False)
```

```python
 df
```

```python
df.shape
```

```python
df["date_added"]=pd.to_datetime(df["date_added"],format="mixed")
df["year_added"]=df["date_added"].dt.year
```

```python
import matplotlib.pyplot as plt
# movies vs ty shows
df["type"].value_counts().plot(kind="bar", color=["red","black"])
plt.title("Movies vs TV shows")
plt.xticks(rotation=45)
plt.savefig("movies_vs_tvshows.png")
plt.show()
```

```python
genres= df["listed_in"].str.split(",").explode()
genres.value_counts().head(10).plot(kind="barh",color="red")
plt.title("Top 10 genres")
plt.grid(True)
plt.tight_layout()
plt.savefig("top_genres.png",bbox_inches="tight")
plt.show()
```

```python
# content added per year
df["year_added"].value_counts().sort_index().plot(kind="line",color="red")
plt.title("Content Added Per Year")
plt.grid(True)
plt.tight_layout()
plt.xlabel("YEARS")
plt.ylabel("movies count")
plt.savefig("content_per_year.png",bbox_inches="tight")
plt.show()
```

```python
# Top 10 countries producing content
df["country"].value_counts().head(10).plot(kind="bar",color="red")
plt.title("Top 10 countries")
plt.xticks(rotation=45)
plt.tight_layout()
plt.ylabel("Movies count")
plt.savefig("top_countries.png",bbox_inches="tight")
plt.show()
```

```python
df["rating"].value_counts().head(10).plot(kind="bar",color="black")
plt.title("Top 10 ratings")
plt.tight_layout()
plt.xticks(rotation=45)
plt.savefig("top_ratings.png",bbox_inches="tight")
plt.show()
```
